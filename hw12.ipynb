```python
import numpy as np
import numpy.linalg as LA  

A = np.array([[2, -1],
              [0, 3]])

det_A = LA.det(A)
print("Determinant of A:", det_A)
A_inv = LA.inv(A)
print("Inverse of A:\n", A_inv)
```

```python
import numpy as np
C = np.array([[4, 1],
              [-2, 1]])

eigenvalues, eigenvectors = np.linalg.eig(C)

print("Matrix C:")
print(C)
print("\nEigenvalue (first):")
print(eigenvalues[0])
print("\nCorresponding Eigenvector:")
print(eigenvectors[:, 0])
```

```python
import numpy as np
X = np.array([[1, 4, 1.5, 100],
              [1, 6, 2.0, 150]])
y = np.array([20, 30])
beta = np.linalg.pinv(X.T @ X) @ X.T @ y
x3 = np.array([1, 8, 3.0, 200])
price_car3 = x3 @ beta
print("Predicted price for Car 3:", price_car3)
```
